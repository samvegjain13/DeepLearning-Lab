```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create Lab 6 folder
os.makedirs("plots_lab6", exist_ok=True)
```

```python
sentence = ["I", "love", "deep", "learning"]
```

```python
attention_weights = np.array([0.1, 0.3, 0.4, 0.2])
```

```python
attention_weights = np.exp(attention_weights) / np.sum(np.exp(attention_weights))
print("Normalized Attention Weights:", attention_weights)
```

```python
plt.figure()

sns.heatmap(
    [attention_weights],
    annot=True,
    xticklabels=sentence,
    cmap="Blues"
)

plt.title("Attention Heatmap")
plt.xlabel("Words")
plt.ylabel("Attention")

# Save in correct folder
plt.savefig("plots_lab6/attention_heatmap.png")

plt.show()
```

```python
sentence = ["This", "movie", "is", "really", "amazing"]
attention_weights = np.array([0.05, 0.1, 0.1, 0.25, 0.5])

attention_weights = np.exp(attention_weights) / np.sum(np.exp(attention_weights))

plt.figure()
sns.heatmap([attention_weights], annot=True, xticklabels=sentence, cmap="Blues")

plt.title("Attention Heatmap (Example 2)")

plt.savefig("plots_lab6/attention_heatmap_2.png")

plt.show()
```

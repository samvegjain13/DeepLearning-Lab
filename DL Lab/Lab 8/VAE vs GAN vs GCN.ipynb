```python
import numpy as np
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
input_dim = 20
latent_dim = 2

inputs = Input(shape=(input_dim,))
encoded = Dense(10, activation='relu')(inputs)
latent = Dense(latent_dim)(encoded)

decoded = Dense(10, activation='relu')(latent)
outputs = Dense(input_dim, activation='sigmoid')(decoded)

vae = Model(inputs, outputs)

vae.compile(optimizer='adam', loss='mse')
vae.summary()
```

```python
data = np.random.rand(1000, input_dim)

vae.fit(data, data, epochs=5, batch_size=32)
```

```python
# Take one sample
sample = data[:1]

# Reconstruct
reconstructed = vae.predict(sample)

print("Original Input:\n", sample)
print("\nReconstructed Output:\n", reconstructed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
generator = Sequential([
    Dense(16, activation='relu', input_shape=(10,)),
    Dense(1, activation='sigmoid')
])
```

```python
discriminator = Sequential([
    Dense(16, activation='relu', input_shape=(1,)),
    Dense(1, activation='sigmoid')
])

discriminator.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
# Fake training demo
noise = np.random.rand(10, 10)
fake_data = generator.predict(noise)

real_data = np.random.rand(10, 1)

X = np.vstack([real_data, fake_data])
y = np.array([1]*10 + [0]*10)

discriminator.train_on_batch(X, y)
```

```python
# Generate new data AFTER training
noise = np.random.rand(5, 10)
generated_data = generator.predict(noise)

print("Generated Data:\n", generated_data)
```
